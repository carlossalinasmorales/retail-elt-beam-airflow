```python
import pandas as pd

data_inputs = "../data/inputs/"

data_santiago = "ventas_santiago.csv"
data_buenos_aires = "ventas_buenos_aires.json"
data_lima = "ventas_lima.parquet"
```

```python
df_santiago = pd.read_csv(data_inputs + data_santiago)
print("**********Info df santiago**********")
print(df_santiago.info())
print("**********Descripcion df santiago**********")
print(df_santiago.describe())
print("**********Tabla df santiago**********")
df_santiago.head(10)
```

```python
df_buenos_aires = pd.read_json(data_inputs + data_buenos_aires, lines=True)
print("**********Info df Buenos Aires**********")
print(df_buenos_aires.info())
print("**********Descripcion df Buenos Aires**********")
print(df_buenos_aires.describe())
print("**********Tabla df Buenos Aires**********")
df_buenos_aires.head(10)
```

```python
df_lima = pd.read_parquet(data_inputs + data_lima, engine="pyarrow")
print("**********Info df lima**********")
print(df_lima.info())
print("**********Descripcion df lima**********")
print(df_lima.describe())
print("**********Tabla df lima**********")
df_lima.head(10)
```
